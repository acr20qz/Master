```python
import pandas as pd
import numpy as np
data = pd.read_csv('task2_data.csv')
data.head()
```

```python
X=data.drop(['y'],axis=1)
y=data.loc[:,'y']
X.head()
y.head()
print(X.shape,y.shape)
```

```python
#创建朴素贝叶斯
from sklearn.naive_bayes import CategoricalNB
model = CategoricalNB()
model.fit(X,y)
```

```python
y_predict_prob=model.predict_proba(X)
print(y_predict_prob)
```

```python
y_predict = model.predict(X)
print(y_predict)
```

```python
#计算模型准确率
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y,y_predict)
print(accuracy)
```

```python
X_test=np.array([[2,1,1,1,1]])
y_test_predict_prob=model.predict_proba(X_test)
y_test_predict=model.predict(X_test)
print(y_test_predict_prob,y_test_predict)
```

```python
X_test=np.array([[2,1,1,1,1],[2,1,1,1,0],[2,1,1,0,0],[2,1,0,0,0],[2,0,0,0,0]])
```

```python
y_test_predict_prob=model.predict_proba(X_test)
y_test_predict=model.predict(X_test)
print(y_test_predict_prob,y_test_predict)
```

```python
#确认存储数据
print(X_test)
print(y_test_predict_prob)
print(y_test_predict)
```

```python
#数据组合

test_result_data=np.concatenate((X_test,y_test_predict_prob,y_test_predict.reshape(5,1)),axis=1)
print(test_result_data)
```

```python
#格式转化
test_result_data=pd.DataFrame(test_result_data)
test_result_data.head()
```

```python
#列名称替换
test_result_data.columns = ['score','school','award','gender','english','p0','p1','p2','y_test_predict']
test_result_data.head()
```

```python
test_result_data.to_csv('test_result_data.csv')
```
