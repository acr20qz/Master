```python
import pandas as pd
import numpy as np
data = pd.read_csv('task2_data.csv')
data.head()
```

```python
from matplotlib import pyplot as plt
```

```python
fig=plt.figure(figsize=(20,5))
fig1=plt.subplot(131)
plt.scatter(data.loc[:,"面积"],data.loc[:,"价格"])
plt.title("Price Vs Size")

fig2=plt.subplot(132)
plt.scatter(data.loc[:,"人均收入"],data.loc[:,"价格"])
plt.title("Price Vs Income")

fig3=plt.subplot(133)
plt.scatter(data.loc[:,"平均房龄"],data.loc[:,"价格"])
plt.title("Price Vs Houseage")

plt.show()
```

```python
#xy赋值
X = data.loc[:,'面积']
y = data.loc[:,'价格']
print(X,y)
```

```python
X=np.array(X)
y=np.array(y)
X=X.reshape(-1,1)
y=y.reshape(-1,1)
print(X.shape,y.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
model.fit(X,y)
```

```python
y_predict=model.predict(X)
print(y_predict)
```

```python
from sklearn.metrics import mean_squared_error,r2_score 
```

```python
MSE = mean_squared_error(y,y_predict)
R2=r2_score(y,y_predict)
print(MSE,R2)
```

```python
fig2=plt.figure(figsize=(8,5))
plt.scatter(data.loc[:,"面积"],data.loc[:,"价格"])
plt.plot(X,y_predict,'r')
plt.title("Price Vs Size")
```

```python
#X,y再次赋值
X=data.drop(['价格'],axis=1)
y=data.loc[:,'价格']
X.head()
```

```python
y=np.array(y)
y=y.reshape(-1,1)
print(X.shape,y.shape)
```

```python
#建立多因子回归模型
model_multi=LinearRegression()
model_multi.fit(X,y)
```

```python
y_predict_multi=model_multi.predict(X)
print(y_predict_multi)
```

```python
MSE = mean_squared_error(y,y_predict_multi)
R2=r2_score(y,y_predict_multi)
print(MSE,R2)
```

```python
fig3=plt.figure(figsize=(8,5))
plt.scatter(y,y_predict_multi)
plt.xlabel('real_y')
plt.ylabel('predict_y')
plt.show()
```

```python
X_test=np.array([[167,70000,5]])
y_test_predict=model_multi.predict(X_test)
print(y_test_predict)
```
