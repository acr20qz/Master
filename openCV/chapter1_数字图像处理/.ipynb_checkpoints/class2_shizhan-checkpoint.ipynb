```python
img_p=io.imread("data/3.jpg")
print(img_p.shape)
plt.imshow(img_p)
```

```python
img_p[100:300,800:900,0]=100
img_p[100:300,800:900,1]=100
img_p[100:300,800:900,2]=100


plt.imshow(img_p)
```

```python
from skimage import draw
h,w=draw.line(1000,200,400,1200)
```

```python
print(h.shape,w.shape)
```

```python
img_p[h,w,0]=255
img_p[h,w,1]=0
img_p[h,w,2]=0

plt.imshow(img_p)


```

```python
draw.set_color(img_p,[h,w],[0,0,255])
plt.imshow(img_p)
```

```python
img_p=io.imread("data/3.jpg")
plt.imshow(img_p)
```

```python
h,w=skimage.draw.circle(200,1000,30)
draw.set_color(img_p,[h,w],[255,100,0])
plt.imshow(img_p)
```

```python
h,w=skimage.draw.circle_perimeter(500,400,30)
draw.set_color(img_p,[h,w],[255,0,0])
plt.imshow(img_p)
```

```python
img_p=io.imread("data/3.jpg")
print(img_p.shape)
from skimage import transform
img_r=transform.resize(img_p,output_shape=[300,300])
print(img_r.shape)
```

```python
plt.imshow(img_r)
```

```python
img_r=transform.rotate(img_p,45,resize=True)
print(img_r.shape)
plt.imshow(img_r)
```

```python
from skimage import exposure
img_g=exposure.adjust_gamma(img_p,0.1)
plt.imshow(img_g)
```
